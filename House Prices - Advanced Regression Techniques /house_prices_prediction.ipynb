### Removing Outliers

#### Get the Numeric and Category Columns

```python
numeric_col_train = [col for col in train_data.columns if train_data[col].dtype == "int64" or train_data[col].dtype == "float64"][1:]
category_cols_train = [col for col in train_data.columns if train_data[col].dtype == "object"]

numeric_col_test = [col for col in test_data.columns if test_data[col].dtype == "int64" or train_data[col].dtype == "float64"][1:]
category_cols_test = [col for col in test_data.columns if test_data[col].dtype == "object"]
```

#### Train Data

```python
before_removal = train_data.shape[0]

for feature in numeric_col_train:
    min_ = train_data[feature].mean() - (3 * train_data[feature].std())
    max_ = train_data[feature].mean() + (3 * train_data[feature].std())
    
    train_data = train_data[train_data[feature] >= min_]
    train_data = train_data[train_data[feature] <= max_]
    
print("Number of Observation Removed: {}".format(before_removal - train_data.shape[0]))
```

## Data Preprocessing

### LabelEncoder()

```python
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()
for feature in category_cols_train:
    train_data[feature] = enc.fit_transform(train_data[feature])
```

#### Test Data

```python
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()
for feature in category_cols_test:
    test_data[feature] = enc.fit_transform(test_data[feature])
```

## Machine Learning Model

### Splitting the Dataset

```python
X = train_data.drop(["Id", "SalePrice"], axis=1).values
y = train_data["SalePrice"].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
y_train
```

### Fitting the Training Data to Model

#### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor().fit(X_train, y_train)
acc_forest = forest.score(X_train, y_train)*100
acc_test_forest = forest.score(X_test, y_test)*100
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gradient = GradientBoostingRegressor().fit(X_train, y_train)
acc_gradient = gradient.score(X_train, y_train)*100
acc_test_gradient = gradient.score(X_test, y_test)*100
```

```python
from sklearn.svm import SVR

svr = SVR(kernel = "rbf").fit(X_train, y_train)
acc_svr = svr.score(X_train, y_train)*100
acc_test_svr = svr.score(X_test, y_test)*100
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor().fit(X_train, y_train)
acc_knn = knn.score(X_train, y_train)*100
acc_test_knn = knn.score(X_test, y_test)*100
```

```python
from sklearn.linear_model import LinearRegression

linearReg = LinearRegression().fit(X_train, y_train)
acc_linearReg = linearReg.score(X_train, y_train)*100
acc_test_linearReg = linearReg.score(X_test, y_test)*100
```

```python
from sklearn.svm import LinearSVR

linearSVR = LinearSVR().fit(X_train, y_train)
acc_linearSVR = linearSVR.score(X_train, y_train)*100
acc_test_linearSVR = linearSVR.score(X_test, y_test)*100
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

tree = DecisionTreeRegressor(max_depth=5, random_state=0).fit(X_train, y_train)
acc_tree = tree.score(X_train, y_train)*100
acc_test_tree = tree.score(X_test, y_test)*100

print("Accuracy on Training Set: {:.2f}".format(acc_tree))
print("Accuracy on Test Set: {:.2f}".format(acc_test_tree))
```

#### Ridge Regression

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1).fit(X_train, y_train)
acc_ridge = tree.score(X_train, y_train)*100
acc_test_ridge = tree.score(X_test, y_test)*100

print("Accuracy on Training Set: {:.2f}".format(acc_ridge))
print("Accuracy on Test Set: {:.2f}".format(acc_test_ridge))
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly_reg = PolynomialFeatures(interaction_only=True)
X_poly = poly_reg.fit_transform(X_train)
reg = LinearRegression()
_ = reg.fit(X_poly, y_train)

acc_poly = reg.score(X_poly, y_train)*100
# acc_test_poly = reg.score(X_train, poly_reg.transform(X_test))*100

print("Accuracy on Training Set: {:.2f}".format(acc_poly))
# print("Accuracy on Training Set: {:.2f}".format(acc_test_poly))
```
