## Data Cleaning

### Inspecting Null Values

```python
# Train Data
train_nulls = train_data.isnull().sum()
train_percent = (train_data.isnull().sum()/train_data.isnull().count())
train_missing_vals = pd.concat([train_nulls, train_percent], axis=1, keys=["Train_Total", "Train_Pecentage"]).sort_values(by="Train_Total", ascending=False)

# Test Data
test_nulls = test_data.isnull().sum()
test_percent = (test_data.isnull().sum()/test_data.isnull().count())
test_missing_vals = pd.concat([test_nulls, test_percent], axis=1, keys=["Test_Total", "Test_Pecentage"]).sort_values(by="Test_Total", ascending=False)

total_null = pd.concat([train_missing_vals, test_missing_vals], axis=1, keys=["Train_Total", "Train_Pecentage", "Test_Total", "Test_Pecentage"])
total_null.columns = total_null.columns.droplevel()
total_null
```

#### Numerical Values

For the numeric values, we will get the mean values and mode values of each column and fill those missing values with that.

It seems only 3 columns does have a null values in numerical column. Let's fill these null values based on their data description.

We can observer  here that the three columns we have filled in train_data (LotFrontage, GarageYrBlt, MasVnrArea)  has missing values here as well, let's copy and paste the code above and deal with this just like before. For the other columns, let's check on what values it holds to fill these values appropriately.

```python
test_data["LotFrontage"].fillna(test_data["LotFrontage"].mean(), inplace=True)
test_data["GarageYrBlt"].fillna(test_data["GarageYrBlt"].median(), inplace=True)
test_data["MasVnrArea"].fillna(test_data["MasVnrArea"].mean(), inplace=True)
```

```python
cols = ["BsmtHalfBath", "BsmtFullBath", "BsmtFinSF2", "GarageCars", "GarageArea", "TotalBsmtSF", "BsmtUnfSF", "BsmtFinSF1"]
for col in cols:
    print(test_data.loc[:10, col].value_counts(dropna=False), "\n")
```

Since some of the columns have values `0.0` (0.0) while some are measurements (Mean) and others are numerical order (Mode), we will handle each column differently

```python
mean_ = ["BsmtFinSF2", "GarageArea", "GarageArea"]
mode_ = ["BsmtHalfBath", "BsmtFullBath"]
zero_ = ["GarageCars", "BsmtUnfSF", "TotalBsmtSF", "BsmtFinSF1"]

for col1 in mean_:
    test_data[col1] = test_data[col1].fillna(test_data[col1].mean())

for col2 in mode_:
    test_data[col2] = test_data[col2].fillna(test_data[col2].mode()[0])
    
for col3 in zero_:
    test_data[col3] = test_data[col3].fillna(0.0)
```

Same procedure for test_data. We checked the columns that have null values and fill those with appropriate values; either mean or mode or a 0.0.

### Removing Outliers

#### Get the Numeric and Category Columns

```python
numeric_col_train = [col for col in train_data.columns if train_data[col].dtype == "int64" or train_data[col].dtype == "float64"][1:]
category_cols_train = [col for col in train_data.columns if train_data[col].dtype == "object"]

numeric_col_test = [col for col in test_data.columns if test_data[col].dtype == "int64" or train_data[col].dtype == "float64"][1:]
category_cols_test = [col for col in test_data.columns if test_data[col].dtype == "object"]
```

#### Train Data

```python
before_removal = train_data.shape[0]

for feature in numeric_col_train:
    min_ = train_data[feature].mean() - (3 * train_data[feature].std())
    max_ = train_data[feature].mean() + (3 * train_data[feature].std())
    
    train_data = train_data[train_data[feature] >= min_]
    train_data = train_data[train_data[feature] <= max_]
    
print("Number of Observation Removed: {}".format(before_removal - train_data.shape[0]))
```

## Data Preprocessing

### LabelEncoder()

#### Train Data

```python
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()
for feature in category_cols_train:
    train_data[feature] = enc.fit_transform(train_data[feature].values)
```

```python
train_data.head()
```

#### Test Data

```python
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()
for feature in category_cols_test:
    test_data[feature] = enc.fit_transform(test_data[feature].values)
```

```python
test_data.head()
```

## Machine Learning Model

### Splitting the Dataset

```python
X = train_data.drop(["Id", "SalePrice"], axis=1).values
y = train_data["SalePrice"].values
```

Here we also removed `Id` from `X` since `Id` is only an identifier and doesn't have anything to do with the independent variable.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
enc = StandardScaler()
# Train Data
X_train = enc.fit_transform(X_train)
# Test Data
X_test = enc.transform(X_test)
```

### Fitting the Training Data to Model

#### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor(max_depth=6, n_estimators=200, n_jobs=-1).fit(X_train, y_train)
acc_forest = forest.score(X_train, y_train)*100
acc_test_forest = forest.score(X_test, y_test)*100

print("Accuracy on Training Set: {:.2f}".format(acc_forest))
print("Accuracy on Test Set: {:.2f}".format(acc_test_forest))
```

#### Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor

gradient = GradientBoostingRegressor(max_depth=1, learning_rate=0.29, max_features=10, random_state=0).fit(X_train, y_train)
acc_gradient = gradient.score(X_train, y_train)*100
acc_test_gradient = gradient.score(X_test, y_test)*100

print("Accuracy on Training Set: {:.2f}".format(acc_gradient))
print("Accuracy on Test Set: {:.2f}".format(acc_test_gradient))
```

```python
from sklearn.svm import SVR

svr = SVR(kernel = "rbf").fit(X_train, y_train)
acc_svr = svr.score(X_train, y_train)*100
acc_test_svr = svr.score(X_test, y_test)*100
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor().fit(X_train, y_train)
acc_knn = knn.score(X_train, y_train)*100
acc_test_knn = knn.score(X_test, y_test)*100
```

```python
from sklearn.linear_model import LinearRegression

linearReg = LinearRegression().fit(X_train, y_train)
acc_linearReg = linearReg.score(X_train, y_train)*100
acc_test_linearReg = linearReg.score(X_test, y_test)*100
```

```python
from sklearn.svm import LinearSVR

linearSVR = LinearSVR().fit(X_train, y_train)
acc_linearSVR = linearSVR.score(X_train, y_train)*100
acc_test_linearSVR = linearSVR.score(X_test, y_test)*100
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

tree = DecisionTreeRegressor(max_depth=5, random_state=0).fit(X_train, y_train)
acc_tree = tree.score(X_train, y_train)*100
acc_test_tree = tree.score(X_test, y_test)*100

print("Accuracy on Training Set: {:.2f}".format(acc_tree))
print("Accuracy on Test Set: {:.2f}".format(acc_test_tree))
```

#### Ridge Regression
